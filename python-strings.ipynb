# sequence sum
1/1! + 2/2! +....+ n/n!

```python
n = int(input("enter n - "))
total = 0

def fact(i):
    res = 1
    for j in range (i, 0, -1):
        res = j * res
    return res
    
for i  in range(1, n+1, 1):
    total += (i/fact(i))
    


print(total)

```

Q. find prime number between given range

```python
l = int(input("enter lowest value : "))
h = int(input("enter highest value : "))

if(l <= 2):
  print(2)

for i in range(max(3,l) ,h+1):
        if(i % 2 == 0):
            continue
        is_prime = True
        for j in range(2, int(i**0.5)+1):
            if(i%j == 0):
                is_prime = False
                break
        if is_prime:
            print(i)
        
```

```python
#capitalise, title,upper, lower swapcase 
s =  "hello world"
print(s.capitalize()) # changes first letter of entire sentence upper
print(s.title()) #changes first letter of every world in upper
print(s.upper()) #changes lower to upper
print(s.lower()) # changes upper to lower
print(s.swapcase()) #change upper to lower and lower to upper
```

```python
# endswith, startswith
s = "it started when i was kid"
print(s.endswith("kid"))
print(s.endswith("kid"))
print(s.startswith("it"))
print(s.startswith("It"))
```

```python
# format  - order matters
s = "it {} when i was {} "
verb = 'started'
obj = 'kid'
print(s.format('started','kid'))
print(s.format(verb,obj))

# format - different order
s1 = "it {1} when i was {0} "
print(s1.format(obj,verb))
```
